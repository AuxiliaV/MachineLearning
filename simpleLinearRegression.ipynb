```python
import pandas as pd     # pandas is a library that handles database and we are going to call pandas as pd
```

```python
#-----------Data Collection----------------------
```

```python
dataset=pd.read_csv("Salary_Data.csv")    # using pd.read_csv, a function in pandas library to read database, we read the "Salary_Data" csv file and assign it to a variable called dataset
```

```python
dataset    #Check whether "Salary_Data" table is assigned to the dataset variable
```

```python
#-----------Input/Output Split----------------------
```

```python
independent=dataset[["YearsExperience"]]    # The whole "YearsExperience" column from the dataset table is assigned to a variable called "independent" which serves as input
```

```python
independent    #Check whether "YearExperience" column is assigned to the independent variable
```

```python
dependent=dataset[["Salary"]]     # The whole "Salary" column from the dataset table is assigned to a variable called "dependent" which serves as output
```

```python
dependent    # Check whether "Salary" column is assigned to the dependent variable
```

```python
#-----------Training and test split----------------------
```

```python
from sklearn.model_selection import train_test_split    # sklearn library supports ML. From sklearn library, under the class "model_selection", we're importing the function called "train_test_split", that splits the input and output data into as set of training data and a set of test data
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)    #We're assigning the original input and output to x_train (training i/p), y_train(training o/p), x_test(test data i/p) y_test(test data o/p) by calling the train_test_split() function and passing the independent and dependent parameters. test_size is the the percentage of the data that has to be test data.
```

```python
#-----------Model Creation----------------------
```

```python
from sklearn.linear_model import LinearRegression    # Importing the LinearRegression function from linear.model class of sklearn library
regressor=LinearRegression()                         # assigning the LinearRegression function to a variable called "regressor"
```

```python
regressor.fit(x_train,y_train)    # We pass the training i/p and o/p "x_train, y_train" parameters to LinearRegression. A model is created
```

```python
weight=regressor.coef_    # Assigning the slope constant value to a variable called "weight"
bias=regressor.intercept_    # Assigning the origin constant value to a variable called "bias"
```

```python
weight    # Check the slope constant value
```

```python
bias    # Check the origin constant value
```

```python
#-----------Evaluate Metrics----------------------
```

```python
y_pred=regressor.predict(x_test)    # Using the predict function of LinearRegression, we're testing the model by passsing the test input values and storing the predicted output values to a variable "y_pred"
```

```python
from sklearn.metrics import r2_score    # Validate the model using r2_score from sklearn library's metric function
r_score=r2_score(y_test,y_pred)    # We're passing the test data o/p (y_test) and predicted data output (y_pred) to check the r2_score and assigning it's value to r_score. The more nearer it is to 1, the better the model performs
r_score
```

```python
#-----------Save the model----------------------
```

```python
import pickle                                     # Using pickle module to save the model
filename="Finalized_model_Linear.sav"             # The "filename" variable is assigned with a string data. this will be the name in which the created model will be saved as
pickle.dump(regressor,open(filename,'wb'))        #pickle.dump saves the code. regressor object will be saved in the name of "Finalized_model_Linear.sav", and the action to be performed is "write binary"
```
